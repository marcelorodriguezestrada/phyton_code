üè∑Ô∏è 21. Generar un grafico donde:
* Argentina tenga formato scatterplot.
* Brasil tenga formato scatterplot+lineplot
* Chile tenga formato lineplot

```
Brazil = df.loc[df.country == "Brazil", :]
Chile = df.loc [df.country == "Chile", :]
```

```
fig = go.Figure()
fig.add_trace(go.Scatter(x = Argentina["year"] , y = Argentina["lifeExp"], mode = "markers", name = "ARG"))
fig.add_trace(go.Scatter(x = Brazil["year"] , y = Brazil["lifeExp"], mode = "lines + markers", name = "BRZ"))
fig.add_trace(go.Scatter(x = Chile["year"] , y = Chile["lifeExp"], mode = "lines", name = "CHL"))

fig.update_layout(
    xaxis_title="PBI per capita",
    yaxis_title="Esperanza de vida"
    )

fig.show()
```

üè∑Ô∏è 22. Modificar el grafico anterior para que el color de la grafica de Argentina es color '#48b051' y un size 15
