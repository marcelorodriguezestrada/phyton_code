üîπ Generar un dataframe que contenga los datos de los vuelos y que contenga una columna con el nombre de la aerolinea.
El nombre de la columna nueva debe ser `aerolinea_nombre`

```python
aerolineas
```

```python
df = vuelos.merge(aerolineas, how = "left", on = "carrier").rename({"name" : "aerolinea_nombre"}, axis = 1)
```

```python
df.head()
```

```python
df
```

üîπ Modificar el dataset generado en el paso anterior para agregar dos columnas que contenga el nombre de la aeropuerto destino y origen.
El nombre de la columna nueva debe ser `aeropuerto_dest` y `aeropuerto_origen`

```python
aeropuerto
```

```python
vuelos
```

```python
df = (vuelos
          #Merge con base aerolineas
          .merge(aerolineas, how = "left", on = "carrier")
          .rename({"name" : "aerolinea_nombre"}, axis = 1)
          #Merge con aeropuertos origen
          .merge(aeropuerto.loc[:, ["faa","name"]],
                 how= "left",
                 left_on = "origin",
                 right_on = "faa")
          .drop("faa", axis = 1)
          .rename({"name" : "aeropuerto_origen"}, axis = 1)
          #Merge con aeropuertos destino
          .merge(aeropuerto.loc[:, ["faa", "name"]],
                 how = "left",
                 left_on ="dest",
                 right_on = "faa")
          .drop("faa", axis = 1)
          .rename({"name" : "aeropuerto_destino"}, axis = 1))
           
```
