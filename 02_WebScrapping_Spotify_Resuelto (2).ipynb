üè∑Ô∏è 1. Importar los paquetes

```
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```
url = 'https://spotifycharts.com/regional'
page = requests.get(url)
```

üè∑Ô∏è 2. Validar el status

```
page.status_code
```

üè∑Ô∏è 3. Imprimir la estructura de la pagina

```
page.text
```

üè∑Ô∏è 4. Hacer un parser del html

```
soup = BeautifulSoup(page.text, 'html.parser')
```

üè∑Ô∏è 5. Generar un objeto que contenga las url de las imagenes

```
imagen=soup.find_all('td',{'class':'chart-table-image'})[0]
```

```
imagen=soup.find_all('td',{'class':'chart-table-image'})
imagen=[i.find_all('img') for i in imagen]
imagen=[i[0]['src'] for i in imagen]
```

üè∑Ô∏è 6. Generar un objeto que contenga el raking

```
ranking=soup.find_all('td',{'class':'chart-table-position'})
ranking=[i.get_text() for i in ranking]
ranking
```

üè∑Ô∏è 7. Generar un objeto que contenga el nombre de la canci√≥n

```
cancion=soup.find_all('td',{'class':'chart-table-track'})
cancion=[i.find_all('strong')[0].get_text().replace('\n','') for i in cancion]
cancion
```

üè∑Ô∏è 8. Generar un objeto que contenga el nombre del autor

```
autor=soup.find_all('td',{'class':'chart-table-track'})
autor=[i.find_all('span')[0].get_text().replace('\n','') for i in autor]
autor
```

üè∑Ô∏è 9. Generar un objeto que contenga el valor de streams

```
streams=soup.find_all('td',{'class':'chart-table-streams'})
streams=[i.get_text() for i in streams]
```

üè∑Ô∏è 10. Generar un dataframe con los objetos generados en los ejercicios anteriores (ranking, canci√≥n, autor, streams e imagen)

```
spotify=pd.DataFrame({'ranking':ranking,'canci√≥n':cancion,'autor':autor,'streams':streams,'imagen':imagen})
spotify
```

üè∑Ô∏è 11. Eliminar el string  `by ` de la columna autor


```
spotify['autor']=[i.replace('by ','') for i in spotify['autor']]
```

```
spotify.head()
```

üè∑Ô∏è 12. Convertir los valores de la columna streams a num√©rico.

```
spotify['streams']=[i.replace(',','') for i in spotify['streams']]
spotify['streams']=spotify['streams'].astype(int)
```

```
spotify.head()
```

üè∑Ô∏è 13. Cuales son los tres autores con m√°s canciones en el ranking y cu√°les son sus canciones.
Las canciones de estos autores ordenarlas en forma ascendente en base a la variable streams



```
temp=list(spotify['autor'].value_counts().unique()[:3])
temp
```

```
autores3p=(
    pd.DataFrame(spotify['autor']
                 .value_counts())
                 .reset_index()
                 .rename({'index':'autor','autor':'qty'},axis=1)
)

autores3p=autores3p.loc[autores3p.qty.isin(temp),:]
autores3p
```

```
spotify.loc[spotify.autor.isin(autores3p.autor),:].sort_values('streams')
```

üè∑Ô∏è 1.Generar la tabla que esta en el siguiente link con las cotizaciones historicas de dolar http://www.dolarhoy.com/DollarChart

```
url = 'http://www.dolarhoy.com/DollarChart'
page = requests.get(url)
```

üè∑Ô∏è 2. Validar el status

```
page.status_code
```

üè∑Ô∏è 3. Imprimir la estructura de la pagina

```
page.text
```

üè∑Ô∏è 4. Hacer un parser del html

```
soup = BeautifulSoup(page.text, 'html.parser')
```

üè∑Ô∏è 5. Generar una tabla que contenga la fecha y las cotizaciones de dolar

```
tabla=soup.find_all('td')
tabla=[i.get_text() for i in tabla]
tabla
```

```
fecha=[]
cotizacion=[]

for i in range(0,len(tabla)):
  if i%2==0:
    fecha.append(tabla[i])
  else:
    cotizacion.append(tabla[i])

coti=pd.DataFrame({'fecha':fecha,'cotizacion':cotizacion})
coti.head()
```

üè∑Ô∏è 6. Convertir en formato datetime la variable `fecha`

```
coti['fecha']=pd.to_datetime(coti['fecha'],format='%m/%d/%Y')
coti.head()
```

üè∑Ô∏è 7. Convertir en float la variale cotizaci√≥n

```
coti['cotizacion']=[i.replace('$','') for i in coti.cotizacion]
coti['cotizacion']=[i.replace(',','.') for i in coti.cotizacion]
coti['cotizacion']=coti['cotizacion'].astype(float)
coti.head()
```
