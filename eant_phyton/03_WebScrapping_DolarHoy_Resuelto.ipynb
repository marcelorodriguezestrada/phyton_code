üè∑Ô∏è 4. Hacer un parser del html

```
soup = BeautifulSoup(page.text, 'html.parser')
```

üè∑Ô∏è 5. Generar una tabla que contenga la fecha y las cotizaciones de dolar

```
tabla=soup.find_all('td')
tabla=[i.get_text() for i in tabla]
tabla
```

```
fecha=[]
cotizacion=[]

for i in range(0,len(tabla)):
  if i%2==0:
    fecha.append(tabla[i])
  else:
    cotizacion.append(tabla[i])

coti=pd.DataFrame({'fecha':fecha,'cotizacion':cotizacion})
coti.head()
```

üè∑Ô∏è 6. Convertir en formato datetime la variable `fecha`

```
coti['fecha']=pd.to_datetime(coti['fecha'],format='%m/%d/%Y')
coti.head()
```

üè∑Ô∏è 7. Convertir en float la variale cotizaci√≥n

```
coti['cotizacion']=[i.replace('$','') for i in coti.cotizacion]
coti['cotizacion']=[i.replace(',','.') for i in coti.cotizacion]
coti['cotizacion']=coti['cotizacion'].astype(float)
coti.head()
```
