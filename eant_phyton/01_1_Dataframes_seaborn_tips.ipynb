üîπ 20. Generar una variable denominada quintil_tip y quintil_total_bill que tenga los valores del 1 al 5 para definir en que quintil esta cada observaci√≥n

```python
m_labels = range(1,6)
```

```python
tips["quintil_tip"]  = pd.qcut(tips["tip"],
        q = 5,
        labels = m_labels )
```

```python
tips["quintil_total_bill"]  = pd.qcut(tips["total_bill"],
        q = 5,
        labels = m_labels )
```
