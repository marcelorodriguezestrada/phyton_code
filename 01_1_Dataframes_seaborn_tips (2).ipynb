üîπ 13. Generemos m√∫ltiples gr√°ficas donde cada una de ellas pueda mostrar los datos relativos a un valor de la dimensi√≥n adicional.
En nuestro caso generemos un scatter plot para cada uno de los üìÖ dias de las semana.
Agregemos los siguientes labels
*   üè∑Ô∏è Titulo: "Grafica en base a la variable day"
*   üè∑Ô∏è Eje x: "Importe del Tickets"
*   üè∑Ô∏è Eje y: "Propinas"

```python
sns.relplot(data = tips,
            y = "tip",
            x ="total_bill", 
            height = 5,
            aspect = 2,
            row="day",
            hue = "sex",
            style = "smoker",
            s = 100,
            palette = sns.color_palette("viridis", n_colors=2),
            kind = "scatter")
plt.title("Mi primer grafico en Python", fontsize = 20)
plt.ylabel('Propinas')
plt.xlabel('Importe Tickets')
```

üîπ 14. Generemos m√∫ltiples gr√°ficas donde cada una de ellas pueda mostrar los datos relativos a un valor de la dimensi√≥n adicional.

```python
sns.relplot(data = tips,
            y = "tip",
            x ="total_bill", 
            row="day",
            col = "smoker",
            hue = "sex",
            s = 100,
            palette = sns.color_palette("viridis", n_colors=2),
            kind = "scatter")
```
